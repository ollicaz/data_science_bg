Criar um gráfico de linhas comparando os números de imigrantes do Brasil e Argentina para o Canadá

```python
df = pd.read_csv('imigrantes_canada.csv')
df.set_index('País', inplace = True) #inplace executa diretamente no dataframe
anos = list(map(str, range(1980, 2014))) #map vai transformar todos os intervalos em string
```

```python
paises = df.loc[['Brasil', 'Argentina'], anos] #Necessário colocar os países em uma lista
paises = paises.T #transposição
paises
```

```python
plt.figure(figsize = (8,4))
plt.plot(paises)
plt.title('Imigração Brasil e Argentina para o Canadá')
plt.legend(['Brasil', 'Argentina'])
plt.xlabel('Ano')
plt.ylabel('Número de imigrantes')
plt.xticks(['1980', '1985', '1990', '1995', '2000', '2005', '2010'])
plt.show()
```

```python
# Durante os primeiros anos houve um crescimento parecido entre os dois países que aumentou a partir dos anos 2000, porém a Argentina decaiu em número de imigrantes e o Brasil aumentou em muito a sua imigração.
```

## Criando figuras com matplotlib

```python
fig, ax = plt.subplots(figsize = (8,4)) #duas variáveis recebendo a mesma função
ax.plot(dados_brasil['ano'], dados_brasil['imigrantes'])
ax.set_title('Imigração do Brasil para o Canadá\n1980 a 2013')
ax.set_xlabel('Ano')
ax.set_ylabel('Número de imigrantes')
ax.xaxis.set_major_locator(plt.MultipleLocator(5))
plt.show()
```

```python
# Boxplot

fig, axs = plt.subplots(1, 2, figsize = (15, 5)) #quantos gráficos queremos colocar na mesma figura

axs[0].plot(dados_brasil['ano'], dados_brasil['imigrantes'])
axs[0].set_title('Imigração do Brasil para o Canadá\n1980 a 2013')
axs[0].set_xlabel('Ano')
axs[0].set_ylabel('Número de imigrantes')
axs[0].xaxis.set_major_locator(plt.MultipleLocator(5))
axs[0].grid()

axs[1].boxplot(dados_brasil['imigrantes']) #boxplot precisa somente dos dados numéricos
axs[1].set_title('Boxplot da imigração do Brasil para o Canadá\n1980 a 2013')
axs[1].set_xlabel('Brasil')
axs[1].set_ylabel('Número de imigrantes')
axs[1].grid()

plt.show()
```

Criando subplots e diversas direções

```python
fig, axs = plt.subplots(2,2, figsize = (10, 6))
fig.subplots_adjust(hspace = 0.5, wspace = 0.3)
fig.suptitle('Imigração dos quatro maiores países da América do Sul para o Canadá\n1980 a 2013')

axs[0,0].plot(df.loc['Brasil', anos])
axs[0,0].set_title('Brasil')

axs[0,1].plot(df.loc['Colômbia', anos])
axs[0,1].set_title('Colômbia')

axs[1,0].plot(df.loc['Argentina', anos])
axs[1,0].set_title('Argentina')

axs[1,1].plot(df.loc['Peru', anos])
axs[1,1].set_title('Peru')

for ax in axs.flat: #loop para percorrer todos os subplots
    ax.xaxis.set_major_locator(plt.MultipleLocator(5))

for ax in axs.flat:
  ax.set_xlabel('Ano')
  ax.set_ylabel('Número de imigrantes')

ymin = 0
ymax = 7000

for ax in axs.ravel(): #transformar um array bidimensional em unidimensional
  ax.set_ylim(ymin, ymax)

plt.show()
```

```python
lojas = ['A', 'B', 'C', 'D']

vendas_2022 = {'Jan': [100, 80, 150, 50],
    'Fev': [120, 90, 170, 60],
    'Mar': [150, 100, 200, 80],
    'Abr': [180, 110, 230, 90],
    'Mai': [220, 190, 350, 200],
    'Jun': [230, 150, 280, 120],
    'Jul': [250, 170, 300, 140],
    'Ago': [260, 180, 310, 150],
    'Set': [240, 160, 290, 130],
    'Out': [220, 140, 270, 110],
    'Nov': [400, 220, 350, 190],
    'Dez': [300, 350, 400, 250]
}
```

```python
vendas_lojas = pd.DataFrame(vendas_2022, index=lojas)
vendas_lojas
```

```python
fig, axs = plt.subplots(2,2, figsize = (18,8))
fig.subplots_adjust(hspace = 0.3, wspace = 0.3)
fig.suptitle('Vendas por lojas 2022', fontsize = 16)

axs[0,0].plot(vendas_lojas.loc['A'])
axs[0,0].set_title('Loja A')

axs[0,1].plot(vendas_lojas.loc['B'])
axs[0,1].set_title('Loja B')

axs[1,0].plot(vendas_lojas.loc['C'])
axs[1,0].set_title('Loja C')

axs[1,1].plot(vendas_lojas.loc['D'])
axs[1,1].set_title('Loja D')

for ax in axs.flat:
  ax.set_xlabel('Mês', fontsize = 12)
  ax.set_ylabel('Vendas', fontsize = 12)

ymin = 0
ymax = 500

for ax in axs.ravel():
  ax.set_ylim(ymin, ymax)

cores = ['darkviolet', 'green', 'darkblue', 'coral']

for i, ax in enumerate(axs.flat): #alterando as cores dos gráficos
  ax.plot(vendas_lojas.loc[vendas_lojas.index[i]], color = cores[i], lw = 2)

plt.show()
```

```python
fig, ax = plt.subplots(figsize = (8,4)) #duas variáveis recebendo a mesma função
ax.plot(dados_brasil['ano'], dados_brasil['imigrantes'], lw = 3, color = 'g')
ax.set_title('Imigração do Brasil para o Canadá\n1980 a 2013', fontsize = 18, loc = 'center')
ax.set_xlabel('Ano', fontsize = 14)
ax.set_ylabel('Número de imigrantes', fontsize = 14)
ax.xaxis.set_tick_params(labelsize = 11)
ax.yaxis.set_tick_params(labelsize = 11)
ax.xaxis.set_major_locator(plt.MultipleLocator(5))

plt.show()
```

```python
america_sul = df.query('Região == "América do Sul"')
america_sul
```

```python
america_sul_sorted = america_sul.sort_values('Total', ascending=True)
```

```python
cores = ['royalblue', 'orange', 'forestgreen', 'orchid', 'purple', 'brown', 'slateblue', 'gray', 'olive', 'navy', 'teal', 'tomato']

fig, ax = plt.subplots(figsize = (12,5))
ax.barh(america_sul_sorted.index, america_sul_sorted['Total'], color = cores )
ax.set_title('Imigração da América do Sul para o Canadá\n1980 a 2013', loc = 'center', fontsize= 14)
ax.set_xlabel('Número de imigrantes', fontsize = 12)
ax.set_ylabel('')
ax.xaxis.set_tick_params(labelsize = 11)
ax.yaxis.set_tick_params(labelsize = 11)
plt.show()
```

```python
cores = []
for pais in america_sul_sorted.index:
  if pais == 'Brasil':
    cores.append('green')
  else:
    cores.append('silver')

fig, ax = plt.subplots(figsize = (12,5))
ax.barh(america_sul_sorted.index, america_sul_sorted['Total'], color = cores )
ax.set_title('América do Sul: Brasil foi o quarto país\n com mais imigrantes para o Canadá\n1980 a 2013', loc = 'center', fontsize= 14)
ax.set_xlabel('Número de imigrantes', fontsize = 12)
ax.set_ylabel('')
ax.xaxis.set_tick_params(labelsize = 11)
ax.yaxis.set_tick_params(labelsize = 11)

for i, v in enumerate(america_sul_sorted['Total']):
  ax.text(v + 20, i, str(v), color = 'black', fontsize = 10, ha = 'left', va = 'center')

ax.set_frame_on(False)
ax.get_xaxis().set_visible(False)
ax.tick_params(axis = 'both', which = 'both', length= 0)

fig.savefig('imigracao_america_sul.png', transparent=False, dpi = 300, bbox_inches = 'tight')

plt.show()
```

## Gráficos com Seaborn

```python
import seaborn as sns
```

```python
sns.set_theme()
```

```python
top_10 = df.sort_values('Total', ascending = False).head(10)
top_10
```

```python
sns.barplot(data = top_10, x = 'Total', y = top_10.index, orient = 'h')
```

```python
ax = sns.barplot(data = top_10, y = top_10.index, x = 'Total', orient = 'h')

ax.set(title = 'Top 10 países com maior imigração para o Canadá\n1980 a 2013',
       xlabel = 'Número de imigrantes',
       ylabel = '')

plt.show()
```

```python
fig, ax = plt.subplots(figsize = (8,4))
ax = sns.barplot(data = top_10, y = top_10.index, x = 'Total', orient = 'h')

ax.set_title('Top 10 países com maior imigração para o Canadá\n1980 a 2013')
ax.set_xlabel('Número de imigrantes', fontsize = 14)
ax.set_ylabel('')
plt.show()
```

```python
def gerar_grafico_paleta(palette):
  fig, ax = plt.subplots(figsize = (8,4))
  ax = sns.barplot(data = top_10, y = top_10.index, x = 'Total', orient= 'h', palette = palette)
  ax.set_title('Top 10 países com maior imigração para o Canadá\n1980 a 2013', fontsize = 15)
  ax.set_xlabel('Número de imigrantes', fontsize = 13)
  ax.set_ylabel('')
  plt.show()
```
